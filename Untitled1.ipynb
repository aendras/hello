```python
from pdf2image import convert_from_path
from transformers import AutoProcessor, Qwen2VLForConditionalGeneration
from mineru_vl_utils import MinerUClient
from PIL import Image

file_path = "main.pdf"

pdf_path = file_path
pages_to_extract = [30, 31, 32]  # zero-based indices for pages 1, 3, 6
poppler_path = r"C:\Users\aendra.shukla\Downloads\Release-23.11.0-0\poppler-23.11.0\Library\bin"

images = convert_from_path(pdf_path, first_page=min(pages_to_extract)+1, last_page=max(pages_to_extract)+1, poppler_path = poppler_path,)

# Filter only the images of the pages you want
selected_images = [images[i - min(pages_to_extract)] for i in pages_to_extract]


# for transformers>=4.56.0
model = Qwen2VLForConditionalGeneration.from_pretrained(
    "opendatalab/MinerU2.5-2509-1.2B",
    dtype="auto", # use `torch_dtype` instead of `dtype` for transformers<4.56.0
    device_map="auto"
)

processor = AutoProcessor.from_pretrained(
    "opendatalab/MinerU2.5-2509-1.2B",
    use_fast=True
)

client = MinerUClient(
    backend="transformers",
    model=model,
    processor=processor
)


results = []

for img in selected_images:
    extracted_blocks = client.two_step_extract(img)
    results.append(extracted_blocks)

# `results` now has parsed data for each chosen page
import json

with open("pdf_extracted.json", "w") as f:
    json.dump(results, f, indent=2)
```

```python
import fitz  # PyMuPDF
from PIL import Image
from transformers import AutoProcessor, Qwen2VLForConditionalGeneration
from mineru_vl_utils import MinerUClient
import json

pdf_path = r"C:\Users\aendra.shukla\mineru\MinerU\main.pdf"
pages_to_extract = [30, 31, 32]

# --- PDF  Images (NO Poppler) ---
doc = fitz.open(pdf_path)

selected_images = []
for p in pages_to_extract:
    page = doc[p]
    pix = page.get_pixmap(dpi=300)
    img = Image.frombytes("RGB", (pix.width, pix.height), pix.samples)
    selected_images.append(img)

# --- MinerU ---
model = Qwen2VLForConditionalGeneration.from_pretrained(
    "opendatalab/MinerU2.5-2509-1.2B",
    device_map="auto",
    dtype="auto"
)

processor = AutoProcessor.from_pretrained(
    "opendatalab/MinerU2.5-2509-1.2B",
    use_fast=True
)

client = MinerUClient(
    backend="transformers",
    model=model,
    processor=processor
)

results = []
for img in selected_images:
    blocks = client.two_step_extract(img)
    results.append(blocks)

with open("pdf_extracted.json", "w") as f:
    json.dump(results, f, indent=2)
```
